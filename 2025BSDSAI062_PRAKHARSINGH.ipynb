```python
# PROBLEM 1
### PART 1:-This function computes the Greatest Common Divisor (GCD) of two integers using the Euclidean Algorithm. The GCD is the largest number that divides both numbers without leaving a remainder.



def gcd(a, b):
    while b!=0:
        remainder=a % b
        a=b
        b=remainder
    return a

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

answer = gcd(num1, num2)

print("The Greatest common divisor is:", answer)
```

```python
### PART 3 :-
def fact(k):
    f=1
    for i in range(1, k+1):           #input should be non negative
        f*=i
    return f

def euler_sum(n):
    s = 0
    for k in range(n):
        s+=1 / fact(k)
    return s
a=euler_sum(0)
print(a)
```

```python
### PART 4 :-
import math

def euler_approx(epsilon):

    sum_val = 0
    k = 0

    while True:

        fact=1
        i=1

        while i <= k:
            fact=fact * i
            i=i + 1

        sum_val = sum_val + (1 / fact)

        if abs(sum_val - math.exp(1)) < epsilon:
            break

        k=k + 1

    return sum_val

a=euler_approx(0.001)
print(a)
```

```python
# PART 5 :-
import math

#euler sum  method
def print_euler_sum_error(n):

    for k in range(1, n+1):

        approx=euler_sum(k)

        error=approx - math.exp(1)

        print(error)

#euler limit method
import math

def print_euler_lim_error(n):

    for k in range(1, n+1):

        approx=euler_limit(k)

        error=approx - math.exp(1)

        print(error)

```

```python
### PROBLEM 3 :- TESTING PROPERTIES OF INTERGERS
# PART 1 :-
def is_power_of_2(n):
    while n > 1:

        if n % 2 != 0:
            return False

        n=n // 2

    return True
is_power_of_2(10)
```
